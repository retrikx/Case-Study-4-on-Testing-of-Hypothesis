# 1. Clarifying whether there is any increase in sales after stepping into digital marketing:

```python
t_value, p_value= stats.ttest_ind(data['Sales_After_digital_add(in $)'], data['Sales_before_digital_add(in $)'], alternative='greater')
```

```python
print(t_value)
print(p_value)
```

[.] Since p-value(=1.3071840034523225e-16) < alpha(=0.05) We reject the null hypothesis . So  there is significant increase in sales after stepping into digital marketing.

# 2. checking whether there is any dependency between the features “Region” and “Manager”.

```python
data['Manager'].value_counts()
```

```python
data['Region'].value_counts()
```

```python
cross=pd.crosstab(data.Manager, data.Region, margins=True)  
cross
```

```python
observed_values=cross.values
observed_values
```

```python
x2, x2_p_value, deg_of_frdm, exp_val = stats.chi2_contingency(observed=observed_values)
```

```python
print(x2)
print(x2_p_value)
print(deg_of_frdm)
print(exp_val)
```

[.] Since p-value(=0.5493991051158094) > alpha(=0.05) We accept the null hypothesis . Here we can see that the two variables, 'Region' and 'Manager' are independent.
